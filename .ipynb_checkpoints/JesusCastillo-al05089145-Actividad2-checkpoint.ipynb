```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm


# Seleccionar las variables
X = df[['bmi']]  # Variable independiente
y = df['charges']  # Variable dependiente

# Agregar constante para el término independiente
X = sm.add_constant(X)

# Ajustar el modelo de regresión lineal
modelo = sm.OLS(y, X).fit()

# Mostrar el resumen del modelo
print(modelo.summary())

# Obtener los coeficientes
intercept, slope = modelo.params
print(f"\nCoeficiente de intersección (intercept): {intercept}")
print(f"Coeficiente de pendiente (slope): {slope}")

# Graficar los datos y la recta de regresión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['bmi'], y=df['charges'], label='Datos')
plt.plot(df['bmi'], intercept + slope * df['bmi'], color='red', label='Recta de regresión')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.title('Regresión lineal simple: Charges vs BMI')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

En este análisis se usó una regresión lineal simple para ver cómo el índice de masa corporal (BMI) influye en los costos del seguro médico. Este resultado es estadísticamente significativo, lo que significa que la relación entre BMI y costo no es casual.<br><br>
Sin embargo, el modelo tiene un <b>R² de 0.039</b>, lo que indica que el BMI solo explica el <b>3.9% de la variación en los costos</b>. En otras palabras, aunque sí existe una relación positiva entre BMI y el costo del seguro, la mayor parte de la variación en los costos se debe a otros factores que no estamos considerando aquí. Entonces, el BMI por sí solo no es suficiente para predecir el costo del seguro con precisión.<br><br>
El valor del intercepto es prácticamente cero <b>(-8.289e-17)</b>, lo cual es normal cuando los datos están estandarizados. El modelo cumple con los requisitos básicos de la regresión, pero para mejorar la predicción sería recomendable agregar más variables al análisis, como edad, si la persona fuma, o el número de hijos.
En resumen, el BMI tiene una influencia positiva y significativa sobre el costo del seguro médico, pero esa influencia es débil si lo analizamos de forma aislada. Para tomar decisiones más acertadas, sería mejor considerar otros factores junto con el BMI.
### Ahora intentemos usar mas variables (Regresión lineal múltiple)

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import statsmodels.api as sm

# Cargar el archivo CSV
df = pd.read_csv("insurance.csv")

# Eliminar valores nulos
df = df.dropna()

# Codificar 'sex' y 'smoker' como variables binarias 
df = pd.get_dummies(df, columns=['sex', 'smoker'], drop_first=True)

print("Tipos de datos por columna")
print(df.dtypes)

#Convertir columnas booleanas a enteros (0 a 1)
for col in df.select_dtypes(include='bool').columns:
    df[col]= df[col].astype(int)
    
print("Tipos de datos por columna, despues de conversión")
print(df.dtypes)

# Codificar el resto de las variables categóricas con LabelEncoder
categorical_columns = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

#Variable dependiente
y = df['charges']

# Seleccionar variables independientes relevantes
#X = df[['age', 'bmi', 'children', 'region', 'sex_male', 'smoker_yes']]
X= df.drop(columns='charges')

# Estandarizar las variables independientes
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Agregar constante
X_scaled = sm.add_constant(X_scaled)

# Ajustar el modelo de regresión lineal múltiple
modelo = sm.OLS(y, X_scaled).fit()

# Mostrar resumen del modelo
print(modelo.summary())

# Mostrar coeficientes
print("\nCoeficientes del modelo:")
print(modelo.params)

# Realizar predicciones
y_pred = modelo.predict(X_scaled)

# Calcular MAE y MSE
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)

# Mostrar resultados de MAE y MSE
print(f"\nError absoluto medio (MAE): {mae:.2f}")
print(f"Error cuadrático medio (MSE): {mse:.2f}")

```

## Interpretación de los coeficientes y métricas del modelo de regresión lineal múltiple

El modelo ajustado tiene un **R² de 0.751**, lo que significa que el 75.1% de la variabilidad en los costos del seguro médico puede explicarse por las variables incluidas. Esto indica un buen poder explicativo y relevancia de las variables seleccionadas.

Además, el desempeño del modelo se evalúa con dos métricas de error:

- **Error absoluto medio (MAE):** 4172.49  
  En promedio, el modelo se equivoca por unos \$4172.49 al predecir el costo del seguro médico de cada persona.

- **Error cuadrático medio (MSE):** 36,527,659.89  
  Este valor representa el promedio de los errores al cuadrado. Un MSE alto indica que hay algunos errores grandes en las predicciones.

### Interpretación de los coeficientes

Cada coeficiente indica el efecto de una variable independiente sobre el costo del seguro médico, considerando que todas las variables han sido estandarizadas (cada coeficiente representa el cambio esperado en el costo por cada aumento de una desviación estándar en la variable correspondiente, manteniendo las demás constantes):

- **x1 (edad):**  
  Por cada aumento de una desviación estándar en la edad, el costo del seguro médico aumenta en promedio 3613.54 unidades. Impacto positivo y significativo.

- **x2 (BMI):**  
  Por cada aumento de una desviación estándar en el índice de masa corporal, el costo aumenta en promedio 2027.32 unidades. Efecto positivo y relevante.

- **x3 (número de hijos):**  
  Por cada aumento de una desviación estándar en el número de hijos, el costo aumenta en promedio 577.66 unidades. Efecto positivo, pero menor.

- **x4 (región):**  
  Por cada aumento de una desviación estándar en la variable región, el costo disminuye en promedio 390.59 unidades. Efecto negativo y pequeño.

- **x5 (sexo):**  
  Por cada aumento de una desviación estándar en la variable sexo, el costo disminuye en promedio 65.55 unidades. Este efecto no es estadísticamente significativo.

- **x6 (fumar):**  
  Por cada aumento de una desviación estándar en la variable fumar, el costo aumenta en promedio 9612.57 unidades. **Este es el efecto más fuerte y significativo del modelo**, lo que confirma que fumar es el factor que más incrementa el costo.

---

### Resumen

El modelo muestra que fumar, la edad y el BMI son los factores más importantes para explicar los costos del seguro médico en este conjunto de datos. El número de hijos y la región tienen efectos menores, y el sexo no parece tener un impacto significativo.  
El modelo es robusto y permite entender cómo cada variable contribuye al costo, siendo fumar el factor de mayor impacto.  
Sin embargo, **los valores de MAE y MSE indican que, aunque el modelo predice bien en general, existen casos donde las predicciones pueden desviarse varios miles de dólares del valor real**, por lo que podría mejorarse incluyendo más variables o ajustando el método.

### Ahora intentemos repetir el ejercicio pero, limpiando los "OUTLIERS"

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import statsmodels.api as sm
import matplotlib.pyplot as plt


df = pd.read_csv("insurance.csv")


df = df.dropna()

#Limpiar outliers en 'charges' usando IQR
Q1 = df['charges'].quantile(0.25)
Q3 = df['charges'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['charges'] >= lower_bound) & (df['charges'] <= upper_bound)]

print(f"Registros eliminados por outliers en 'charges': {1399 - len(df)}")
print("\nNuevo resumen estadístico de 'charges':")
print(df['charges'].describe())
print(f"\nTamaño final del dataset limpio: {len(df)}")

#Codificar 'sex' y 'smoker' como variables binarias 
df = pd.get_dummies(df, columns=['sex', 'smoker'], drop_first=True)

print("Tipos de datos por columna")
print(df.dtypes)

#Convertir columnas booleanas a enteros (0 a 1)
for col in df.select_dtypes(include='bool').columns:
    df[col]= df[col].astype(int)
    
print("Tipos de datos por columna, después de conversión")
print(df.dtypes)

#Codificar el resto de las variables categóricas con LabelEncoder
categorical_columns = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

#Variable dependiente
y = df['charges']

#Seleccionar variables independientes relevantes
X = df.drop(columns='charges')

#Estandarizar las variables independientes
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Agregar constante
X_scaled = sm.add_constant(X_scaled)

#Ajustar el modelo de regresión lineal múltiple
modelo = sm.OLS(y, X_scaled).fit()

#Mostrar resumen del modelo
print(modelo.summary())

#Mostrar coeficientes
print("\nCoeficientes del modelo:")
print(modelo.params)

#Realizar predicciones
y_pred = modelo.predict(X_scaled)

#Calcular MAE y MSE
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)

#Mostrar resultados de MAE y MSE
print(f"\nError absoluto medio (MAE): {mae:.2f}")
print(f"Error cuadrático medio (MSE): {mse:.2f}")



```

### Ahora realizaremos una "Regresion Logística"

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import statsmodels.api as sm

# Cargar datos
df = pd.read_csv("insurance.csv")
df = df.dropna()

# Codificar variables categóricas
df_encoded = df.copy()
for col in ['sex', 'smoker', 'region']:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Crear variable binaria, donde el costo es alto si high_cost=1, y bajo costo (low cost) si high_cost=0
mean_charge = df_encoded['charges'].mean()
df_encoded['high_cost'] = (df_encoded['charges'] > mean_charge).astype(int)

# Seleccionar predictores
X = df_encoded[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
X = sm.add_constant(X)  # Agregar constante para el modelo
y = df_encoded['high_cost']

# Modelo Logit (regresión logística binaria)
model_bin = sm.Logit(y, X)
result_bin = model_bin.fit()
print(result_bin.summary())


```

## Interpretación del modelo logit para clasificar clientes como "alto costo" o "bajo costo"

El modelo logit estima la probabilidad de que un cliente tenga gastos médicos por encima del promedio ("alto costo"). Los resultados muestran que **la edad** y **ser fumador** son los factores más importantes:

- **Ser fumador tiene un impacto enorme**: el coeficiente es **8.39**, lo que indica que los fumadores tienen una probabilidad muchísimo mayor de estar en el grupo de alto costo comparado con los no fumadores.
- Por cada año adicional de edad, la probabilidad de ser "alto costo" aumenta, con un coeficiente de **0.0709**.

Por otro lado, el sexo (**-0.2744**), el índice de masa corporal (**0.0213**) y el número de hijos (**0.1203**) no muestran una influencia estadísticamente significativa en este modelo, ya que sus valores p son mayores a 0.05. Es decir, no hay evidencia suficiente para decir que estos factores cambian la probabilidad de ser "alto costo" en este conjunto de datos. 

La región donde vive el cliente también influye, aunque menos: el coeficiente es **-0.1995**, lo que significa que hay regiones donde es menos probable ser "alto costo". Esto porque la región tiene un valor p 0.020 lo cual es menor que 0.05, lo cual significa que su impacto es significativo. El valor negativo indica que si la numeración de las regiones fue 1,2,3.. a medida que crecía el alto costo disminuye.

El modelo en general es muy bueno para distinguir entre clientes de alto y bajo costo, ya que el **Pseudo R-cuadrado es 0.54**, lo que indica que explica el 54% de la variabilidad en la clasificación. Es decir que este modelo mas o menos puede predecir correctamente en la mitad de los casos.

**En resumen:**  
Los clientes más propensos a tener **costos altos son los fumadores y las personas mayores**. Los **no fumadores y los jóvenes** tienen más posibilidades de ser **"bajo costo"**. La región si influye en el costo pero su influencia es menor. 
El sexo, el IMC y el número de hijos no parecen ser determinantes en este caso. 

## Ahora generemos predicciones y evaluemos la precisión usando una matriz de confusión y métricas como precision, recall y F1 Score.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
import statsmodels.api as sm

# Cargar los datos
df = pd.read_csv("insurance.csv")
df = df.dropna()

# Codificar variables categóricas
df_encoded = df.copy()
for col in ['sex', 'smoker', 'region']:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Crear variable binaria: high_cost = 1 si charges > media, 0 en caso contrario
mean_charge = df_encoded['charges'].mean()
df_encoded['high_cost'] = (df_encoded['charges'] > mean_charge).astype(int)

# Seleccionar predictores y variable objetivo
X = df_encoded[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
X = sm.add_constant(X)
y = df_encoded['high_cost']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Ajustar modelo logit
model = sm.Logit(y_train, X_train)
result = model.fit()

# Predecir probabilidades y convertir a clases (umbral 0.5)
y_pred_prob = result.predict(X_test)
y_pred = (y_pred_prob >= 0.5).astype(int)

# Obtener matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
TN, FP, FN, TP = conf_matrix.ravel()

print("\nMatriz de confusión")
print(f" Verdaderos negativos (TN): {TN}")
print(f" Falsos positivos (FP): {FP}")
print(f" Falsos negativos (FN): {FN}")
print(f" Verdaderos positivos (TP): {TP}")

# Métricas de precisión, recall y F1 Score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"\nPrecisión (alto costo): {precision:.2f}")
print(f"Recall (alto costo): {recall:.2f}")
print(f"F1 Score (alto costo): {f1:.2f}")

# También puedes imprimir el reporte completo
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))


```

## Evaluación de precisión usando una matriz de confusión y métricas como precision, recall y F1 Score
**Matriz de confusión**

## Apliquemos ahora una regresión logística multinomial y ordinal, asi mismo usaremos una matriz de confusión y métricas como precision, recall y F1 Score para medir

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Cargar los datos
df = pd.read_csv("insurance.csv")

# Codificar variables categóricas
for col in ['sex', 'smoker', 'region']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Crear variable categórica de tres niveles según terciles
terciles = df['charges'].quantile([0.33, 0.66])
def categorize_charge(charge):
    if charge <= terciles[0.33]:
        return 'bajo'
    elif charge <= terciles[0.66]:
        return 'medio'
    else:
        return 'alto'

df['cost_category'] = df['charges'].apply(categorize_charge)
df['cost_category_num'] = df['cost_category'].map({'bajo': 0, 'medio': 1, 'alto': 2})

# Seleccionar predictores y variable objetivo
X = df[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
y = df['cost_category_num']

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Regresión logística multinomial
model = LogisticRegression( solver='lbfgs', max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['bajo', 'medio', 'alto'],
            yticklabels=['bajo', 'medio', 'alto'])
plt.xlabel('Predicción')
plt.ylabel('Valor real')
plt.title('Matriz de Confusión - Regresión Logística Multinomial')
plt.tight_layout()
plt.show()

# Reporte de clasificación
report = classification_report(y_test, y_pred, target_names=['bajo', 'medio', 'alto'])
print("Reporte de clasificación:\n")
print(report)
```

### Clientes de bajo costo
Para los clientes de bajo costo, el modelo tiene una precisión de ***0.90***, lo que significa que el 90% de los que predice como bajo realmente lo son, y un recall de ***0.92***, es decir, identifica correctamente al 92% de todos los clientes que realmente son de bajo costo. El F1 Score para este grupo es 0.91, lo que indica un buen equilibrio entre precisión y recall.

### Clientes de costo medio
En el grupo de costo medio, la precisión es de ***0.71***, así que el 71% de los que predice como medio realmente pertenecen a esa categoría. El recall es muy alto,***0.96***, lo que significa que el modelo logra encontrar casi todos los clientes de costo medio, aunque a veces confunde otros grupos con este. El F1 Score es ***0.82*** el cual tambien es un buen equilibrio entre precisión y recall.

### Clientes de costo alto
Para los clientes de alto costo, la precisión es de ***0.98***, lo que indica que casi todos los que predice como alto realmente lo son, pero el recall es de ***0.62***, así que solo identifica correctamente al 62% de todos los clientes que realmente son de alto costo. El F1 Score para este grupo es ***0.76***.
La exactitud global del modelo es de 0.84, lo que significa que el 84% de las predicciones son correctas. En promedio, el modelo tiene una precisión de 0.86, un recall de 0.83 y un ***F1 Score de 0.83*** considerando las tres clases. Esto indica que el modelo es bueno para distinguir entre los tres niveles de costo, aunque tiene más dificultad para encontrar todos los clientes de alto costo.

### Ahora incluyamos una Regresión logística ordinal, repitiendo los mismos pasos anteriores y comparemos

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.miscmodels.ordinal_model import OrderedModel

# Cargar los datos
df = pd.read_csv("insurance.csv")
df = df.dropna()

# Codificar variables categóricas
for col in ['sex', 'smoker', 'region']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Crear variable categórica de tres niveles según terciles
terciles = df['charges'].quantile([0.33, 0.66])
def categorize_charge(charge):
    if charge <= terciles[0.33]:
        return 'bajo'
    elif charge <= terciles[0.66]:
        return 'medio'
    else:
        return 'alto'

df['cost_category'] = df['charges'].apply(categorize_charge)
df['cost_category_num'] = df['cost_category'].map({'bajo': 0, 'medio': 1, 'alto': 2})

# Seleccionar predictores y variable objetivo
X = df[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
y = df['cost_category_num']

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo Multinomial
multi_model = LogisticRegression(solver='lbfgs', max_iter=1000)
multi_model.fit(X_train, y_train)
y_pred_multi = multi_model.predict(X_test)

# Matriz de confusión multinomial
conf_matrix_multi = confusion_matrix(y_test, y_pred_multi)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix_multi, annot=True, fmt='d', cmap='Blues',
            xticklabels=['bajo', 'medio', 'alto'],
            yticklabels=['bajo', 'medio', 'alto'])
plt.xlabel('Predicción Multinomial')
plt.ylabel('Valor real')
plt.title('Matriz de Confusión - Multinomial')
plt.show()

print("Reporte de clasificación (Multinomial):\n")
print(classification_report(y_test, y_pred_multi, target_names=['bajo', 'medio', 'alto']))
print("Accuracy Multinomial:", accuracy_score(y_test, y_pred_multi))

# Modelo Ordinal
ord_model = OrderedModel(y_train, X_train, distr='logit')
ord_result = ord_model.fit(method='bfgs')
y_pred_ord = ord_result.predict(X_test).idxmax(axis=1)

# Matriz de confusión ordinal
conf_matrix_ord = confusion_matrix(y_test, y_pred_ord)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix_ord, annot=True, fmt='d', cmap='Greens',
            xticklabels=['bajo', 'medio', 'alto'],
            yticklabels=['bajo', 'medio', 'alto'])
plt.xlabel('Predicción Ordinal')
plt.ylabel('Valor real')
plt.title('Matriz de Confusión - Ordinal')
plt.show()

print("Reporte de clasificación (Ordinal):\n")
print(classification_report(y_test, y_pred_ord, target_names=['bajo', 'medio', 'alto']))
print("Accuracy Ordinal:", accuracy_score(y_test, y_pred_ord))
```

### Comparación de modelos: Multinomial vs Ordinal
> Clase Bajo Costo

> - ***Multinomial***: Precisión 0.90, recall 0.92, F1-score 0.91.
> - ***Ordinal***: Precisión 0.90, recall 0.91, F1-score 0.86.

> Clase Medio Costo

> - ***Multinomial***: Precisión 0.71, recall 0.96, F1-score 0.82.
> - ***Ordinal***: Precisión 0.82, recall 0.92, F1-score 0.86.


> Clase Alto Costo

> - ***Multinomial***: Precisión 0.98, recall 0.62, F1-score 0.76.
> - ***Ordinal***: Precisión 0.93, recall 0.81, F1-score 0.86.

### Conclusiones:
Ambos modelos tienen un desempeño muy similar, con precisión y recall altos.
El multinomial tiene un F1-score ligeramente superior (0.91 vs 0.86), lo que indica un mejor equilibrio entre precisión y recall en esta clase.
Para identificar clientes de bajo costo, cualquiera de los dos modelos es adecuado, aunque el multinomial tiene una ligera ventaja en el balance general.

### Ahora pasemos a ejecutar una Regresión polinomial de grado 2, 3 y 5 entre bmi y charges.

```python
import pandas as pd, numpy as np, statsmodels.api as sm, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# --- Habilitar visualización inline en Jupyter (seguro si no estás en Jupyter) ---
try:
    from IPython import get_ipython
    ipy = get_ipython()
    if ipy is not None:
        ipy.run_line_magic('matplotlib', 'inline')
        import matplotlib
        matplotlib.rcParams['figure.dpi'] = 130  # más nítido en notebook
except Exception:
    pass

# Carga y limpieza
df = pd.read_csv('insurance.csv')
df = df.dropna()

df.columns = [c.strip() for c in df.columns]     # limpiar posibles espacios
X = df[['bmi']].copy()
y = df['charges'].copy()
X = X.dropna()
y = y.loc[X.index]

# Train / Test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# Ajuste polinomial y métricas
grados = [2, 3, 5]
registros = []
plot_paths = []

for grado in grados:
    # --- Matrices de diseño con intercepto e iterando potencias de BMI ---
    Xtr = pd.DataFrame(index=X_train.index)
    Xte = pd.DataFrame(index=X_test.index)
    Xtr['const'] = 1.0
    Xte['const'] = 1.0
    for d in range(1, grado + 1):
        Xtr[f'bmi^{d}'] = X_train['bmi'] ** d
        Xte[f'bmi^{d}'] = X_test['bmi'] ** d

    # 4) Ajuste OLS con statsmodels
    modelo = sm.OLS(y_train, Xtr).fit()

    # 5) Predicciones
    yhat_tr = modelo.predict(Xtr)
    yhat_te = modelo.predict(Xte)

    # 6) Métricas (train)
    R2_tr        = modelo.rsquared
    R2_adj_tr    = modelo.rsquared_adj
    MAE_tr       = mean_absolute_error(y_train, yhat_tr)
    MSE_tr       = mean_squared_error(y_train, yhat_tr)

    # 7) Métricas (test)
    R2_te        = r2_score(y_test, yhat_te)
    n_te         = len(y_test)
    p            = grado  # número de regresores sin contar el intercepto
    # R² ajustado en test (referencial; no lo reporta statsmodels)
    R2_adj_te    = 1 - (1 - R2_te) * (n_te - 1) / (n_te - p - 1)
    MAE_te       = mean_absolute_error(y_test, yhat_te)
    MSE_te       = mean_squared_error(y_test, yhat_te)

    registros.append({
        'degree': grado,
        'R2_train': R2_tr, 'R2_adj_train': R2_adj_tr, 'MAE_train': MAE_tr, 'MSE_train': MSE_tr,
        'R2_test': R2_te, 'R2_adj_test': R2_adj_te, 'MAE_test': MAE_te, 'MSE_test': MSE_te
    })

    # 8) Curva ajustada sobre rejilla para graficar
    bmi_min, bmi_max = X['bmi'].min(), X['bmi'].max()
    bmi_grid = np.linspace(bmi_min, bmi_max, 400)
    Xg = pd.DataFrame({'const': np.ones_like(bmi_grid)})
    for d in range(1, grado + 1):
        Xg[f'bmi^{d}'] = bmi_grid ** d
    y_grid = modelo.predict(Xg)

    plt.figure(figsize=(7, 5))
    plt.scatter(X_train['bmi'], y_train, s=12, alpha=0.5, label='Train')
    plt.scatter(X_test['bmi'], y_test,   s=12, alpha=0.5, label='Test')
    plt.plot(bmi_grid, y_grid, color='black', lw=2, label=f'Poly deg {grado}')
    plt.xlabel('BMI'); plt.ylabel('Charges')
    plt.title(f'Polynomial Regression (degree={grado})\nTrain R²={R2_tr:.3f}, Test R²={R2_te:.3f}')
    plt.legend(); plt.tight_layout()
    plt.show()                   
        

# Seleccionar el “mejor” por R² de test y graficar residuales (train)
res_df = pd.DataFrame(registros)
mejor_grado = res_df.loc[res_df['R2_test'].idxmax(), 'degree']

# reconstruir diseño para mejor_grado (train) para residuales
Xtr_best = pd.DataFrame(index=X_train.index)
Xtr_best['const'] = 1.0
for d in range(1, int(mejor_grado) + 1):
    Xtr_best[f'bmi^{d}'] = X_train['bmi'] ** d

modelo_best = sm.OLS(y_train, Xtr_best).fit()
yhat_tr_best = modelo_best.predict(Xtr_best)
residuales = y_train - yhat_tr_best

plt.figure(figsize=(7,5))
plt.scatter(yhat_tr_best, residuales, s=12, alpha=0.6)
plt.axhline(0, color='red', ls='--')
plt.xlabel('Fitted values (train)'); plt.ylabel('Residuals')
plt.title(f'Residuals vs Fitted (degree={int(mejor_grado)})')
plt.tight_layout()
plt.show()                                   


# 10) Exportar tabla de métricas a CSV y mostrar tablas redondeadas
res_df_rounded = res_df.copy()
for col in res_df_rounded.columns:
    if col != 'degree':
        res_df_rounded[col] = res_df_rounded[col].astype(float).round(4)

res_df_rounded.to_csv('poly_metrics_train_test.csv', index=False)

print('== Métricas (train) ==')
print(res_df_rounded[['degree','R2_train','R2_adj_train','MAE_train','MSE_train']])
print('\n== Métricas (test) ==')
print(res_df_rounded[['degree','R2_test','R2_adj_test','MAE_test','MSE_test']])
print('CSV métricas: poly_metrics_train_test.csv')
```

|   3   | 0.02    | 9,863   | 152,453,000 | 0.01         |

- **¿Qué significa R^2?**  
  Es el porcentaje de la variación de los costos (charges) que el modelo logra explicar usando solo el índice de masa corporal (bmi).

- **¿Qué muestran los resultados?**
  - **Todos los modelos explican muy poco** (solo 2–3%).
  - **El error promedio (MAE)** es casi igual en todos: alrededor de 9,800.
  - **El error cuadrático medio (MSE)** también es muy alto y similar.
  - **Aumentar el grado del polinomio NO mejora el modelo**; de hecho, lo empeora ligeramente.

### ¿Hay sobreajuste?

Sí. Al aumentar el grado del polinomio (de 2 a 5), el modelo empeora en datos nuevos:

- **R^2 en test baja** de 0.03 a 0.02
- **MSE (error) en test sube** de 150,472,800 a 152,679,200

Esto indica que el modelo se ajusta demasiado al entrenamiento y pierde capacidad de generalizar.

---

### Conclusiones

- **Aumentar el grado del polinomio no ayuda y puede empeorar la predicción en datos nuevos.**

### Ahora apliquemos Ridge y Lasso sobre las mismas variables de la regresión múltiple

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.model_selection import train_test_split

try:
    from IPython import get_ipython
    ipy = get_ipython()
    if ipy is not None:
        ipy.run_line_magic('matplotlib', 'inline')
        import matplotlib
        matplotlib.rcParams['figure.dpi'] = 130
except Exception:
    pass

df = pd.read_csv("insurance.csv").dropna()

# Codificar 'sex' y 'smoker' como dummies binarios (drop_first=True para evitar colinealidad)
df = pd.get_dummies(df, columns=['sex', 'smoker'], drop_first=True)

# Asegurar enteros en columnas booleanas (si las hubiera)
for col in df.select_dtypes(include='bool').columns:
    df[col] = df[col].astype(int)

# LabelEncoder para las categóricas restantes (p. ej., 'region')
categorical_columns = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Variable objetivo y predictores (mismas variables de la múltiple: todo menos 'charges')
y = df['charges'].values
X = df.drop(columns='charges')

# 2) Train/Test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# 3) Escalado (necesario para Ridge/Lasso; OK también para Lineal)
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

# 4) Definir modelos
models = {
    "Lineal": LinearRegression(),
    "Ridge": Ridge(alpha=1.0, random_state=None),
    "Lasso": Lasso(alpha=0.1, random_state=None, max_iter=10000)
}

# 5) Entrenar, predecir y evaluar
rows = []
y_preds_test = {}  # para graficar
for name, mdl in models.items():
    mdl.fit(X_train_sc, y_train)

    # Predicciones
    yhat_tr = mdl.predict(X_train_sc)
    yhat_te = mdl.predict(X_test_sc)
    y_preds_test[name] = yhat_te

    # Métricas
    r2_tr  = r2_score(y_train, yhat_tr)
    r2_te  = r2_score(y_test, yhat_te)

    mae_tr = mean_absolute_error(y_train, yhat_tr)
    mae_te = mean_absolute_error(y_test, yhat_te)

    mse_tr = mean_squared_error(y_train, yhat_tr)
    mse_te = mean_squared_error(y_test, yhat_te)

    # Variables activas (coeficientes != 0)
    # Nota: Linear y Ridge raramente ponen coeficientes exactamente en 0
    n_act  = int(np.sum(np.abs(mdl.coef_) > 1e-12))

    rows.append({
        "Modelo": name,
        "R2_train": r2_tr, "MAE_train": mae_tr, "MSE_train": mse_tr,
        "R2_test":  r2_te, "MAE_test":  mae_te, "MSE_test":  mse_te,
        "Vars_activas": n_act
    })

# 6) Tabla comparativa
tabla = pd.DataFrame(rows)
tabla_round = tabla.copy()
for c in ["R2_train","R2_test"]:
    tabla_round[c] = tabla_round[c].round(4)
for c in ["MAE_train","MAE_test","MSE_train","MSE_test"]:
    tabla_round[c] = tabla_round[c].round(2)

print("=== Comparativa (train/test) ===")
print(tabla_round)

# 7) Visualización: Predicción vs Valor Real (en TEST) para cada modelo
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharex=True, sharey=True)
axes = np.ravel(axes)

y_min = min(y_test.min(), *[y_preds_test[m].min() for m in models])
y_max = max(y_test.max(), *[y_preds_test[m].max() for m in models])

for ax, (name, yhat) in zip(axes, y_preds_test.items()):
    ax.scatter(y_test, yhat, s=15, alpha=0.6, edgecolor='none')
    ax.plot([y_min, y_max], [y_min, y_max], 'r--', lw=1)  # línea y = x
    # buscar métricas de este modelo para anotarlas
    row = tabla_round[tabla_round["Modelo"] == name].iloc[0]
    ax.set_title(f"{name}\nR²={row['R2_test']}, MAE={row['MAE_test']}")
    ax.set_xlabel("Valor real (charges)")
    ax.set_ylabel("Predicción (charges)")

plt.tight_layout()
plt.show()

# 8) (Opcional) También puedes guardar la tabla a CSV
# tabla_round.to_csv("comparativa_lineal_ridge_lasso.csv", index=False)
```
